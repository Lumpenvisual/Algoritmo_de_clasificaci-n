````python
#ALGORITMO o scrip DE CLASIFICACIÓN =tema base cancer de mama


#librerias esenciales para hacer inteligencia artificial con Python
#para analis estadisticos, sumatoria, graficas,
#modelo de clasificación clasificar un modelo para cancer de mama a nivel real
#hay muchos algoritmos para machine learning y uno de los más importantes es este
#de clasificación
#hacer un modelo de trading no es tan fácil depende de sentimiento en vivo a nivel mundial
#de cuando va a caer una acción
#se basa en los parámetros de información de las redes de datos, se necesita velocidad dedicada
#en internet
#en criptos depende del sistema de gas y tokens
#probabilidad de anális deportivo hay que sumar muchas variables
#se pueden hacer clustering cajones para almacenar información para que la IA
#se enfrente sola al mundo, por ejemplo Meta
#hay datas que son cerradas y hay que pedir información para acceder a ellas
#sacar población de desplazamiento hay que acceder a la división política de colombia
#acá tenemos distribucíon de clases, clasificación, matris de graficación
#algorismos de ramdom forest
#sbm de maquina de vectores
#xybox
#regrsión lineal simple
#algoritmo polinómico


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#----------------------------
#modelos de datos: analisis de datos de tendencias de una ciudad, ventas, clasificacion de
#asesores, sistemas de dinero, sistemas sanitarios, mercados potentes

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
#ejes de comando de dónde saco los datos
#sklearn (load breast cancer) es de donde se saca la data
from sklearn.datasets import load_breast_cancer

#------------------
#configurar motor de estilos gráficos
plt.style.use('ggplot')

#conecto el matplilib y lo conecto en linea
%matplotlib inline

#------------------------

# 1. cargar y explorar los datos
#tratar que los datos no sean tan planos y que sean más graficos

print("======1. Cargando y explorando los datos ===")
data = load_breast_cancer()
df=pd.DataFrame(data.data, columns=data.feature_names)
#hacer el binario y convertir dato a 0 y 1
df['target'] = data.target
print("\nPrimeras 5 filas de dataset:")
print(df.head())
#enviar la información clasificando las variables, luego se puede clasificar por
#periodos pequeños o rangos específicos

print ("\nInformación del dataset")
print(df.info())
print('\nEstadisticas descriptivas')
print(df.describe())

#visualización de los datos, analítica descriptiva visual
print('\n===2. Visualización de datos===')
plt.figure(figsize=(12,8))
sns.countplot(x='target', data=df)
plt.title('Distribucción de casos (0: Maligno, 1: Benogno)')
plt.show()


#Histogramas de las caracteristicas
df.drop('target', axis=1).hist(bins=20, figsize=(20, 15))
plt.suptitle('Distribución de características', y=1.02)
plt.show()

#Matriz de colerración
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title('Matriz de correlación de caracteristicas')
plt.show()

#preprocesamiento de datos para darle contexto
print('\n===3. Preprocesamiento de datos===')
X = df.drop('target', axis=1)
y = df['target']
#Dividir en conjuntos de entrenamiento y prueba para clasificarlos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Escalar caracteristicas
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print("===SALIDAS===")
print(f"\nForma de los datos de entrenamiento: {X_train.shape}")
print(f"Forma de los datos de prueba: {X_test.shape}")

#Entrenamiento del modelo:divirgencia como debe comportarse el algoritmo
print('\n===4. Entrenamiento del modelo===')
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

#Evaluación del modelo
print('\n===5. Evaluación del modelo===')
y_pred = model.predict(X_test)
print("\n==Reportede clasificación==:")
print(classification_report(y_test, y_pred))

#------------------------------------
#Las matrices de confusión son herramientas esenciales en el
#aprendizaje automático, especialmente para evaluar el rendimiento
#de modelos de clasificación. permiten visualizar y analizar la precisión
#de las predicciones, indicando errores comunes y áreas donde
#el modelo necesita mejoras

print("\nMatriz de confusión:")
#mapeador
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de confusión')
plt.xlabel('Predictivo')
plt.ylabel('Real')
plt.show

#precision al modelo
print(f"\nPrecisión del modelo:{accuracy_score(y_test,y_pred):.2f}" )
#importancia de las características
print("\===6. Importancia de las características====")
feature_importance = pd.Series(model.feature_importances_,index=data.feature_names)
feature_importance.nlargest(10).plot(kind='barh', figsize=(10,8))
plt.title('Importancia de las características')
plt.show()

#muestreo y modelo de predicción de patrones
print('\n===7. Predicción de ejemplo ===')
sample = X_test[0].reshape(1,-1)
prediccion=model.predict(sample)
print(f"\nPredicción para muestra de ejemplo o de prueba: {'Benigno' if prediccion[0] == 1 else 'Malingo'}")
print(f"Etiqueta real: {'Benigno' if y_test.iloc[0] == 1 else 'Maligno'}")
````
